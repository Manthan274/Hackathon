```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style='darkgrid')
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

label=LabelEncoder()
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df['Dependents'] = df['Dependents'].str.rstrip('+')
df['Gender'] = df['Gender'].fillna( df['Gender'].dropna().mode().values[0] )
df['Married'] = df['Married'].fillna( df['Married'].dropna().mode().values[0] )
df['Dependents'] = df['Dependents'].fillna( df['Dependents'].dropna().mode().values[0] )
df['Self_Employed'] = df['Self_Employed'].fillna( df['Self_Employed'].dropna().mode().values[0] )
df['LoanAmount'] = df['LoanAmount'].fillna( df['LoanAmount'].dropna().mean() )
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna( df['Loan_Amount_Term'].dropna().mode().values[0] )
df['Credit_History'] = df['Credit_History'].fillna( df['Credit_History'].dropna().mode().values[0] )

df['Loan_Status'] = df['Loan_Status'].map({'N':0, 'Y':1}).astype(np.int)

df.reset_index(drop=True, inplace=True)

for i in  range(len(df)):
    if df['CoapplicantIncome'][i]>0:
        df['CoapplicantIncome'][i]=1

df.reset_index(drop=True, inplace=True)
df.info()
```

```python
list=['Gender','Married','Education','Self_Employed','Self_Employed']
for i in list:
    df[i]=label.fit_transform(df[i]).astype(int)

columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Loan_Status']

df=df[columns]
df.reset_index(drop=True, inplace=True)
print(df.info())
```

```python
X, y  = df.iloc[:, 0:-1], df.iloc[:, -1]
X= pd.get_dummies(X)
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=62)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
slc= StandardScaler()
X_train_std = slc.fit_transform(X_train)
X_test_std = slc.fit_transform(X_test)

X=slc.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators =500, criterion='entropy', oob_score=True, random_state=1,n_jobs=-1)
```

```python
from xgboost.sklearn import XGBClassifier
xgb1 = XGBClassifier(learning_rate =0.1, n_estimators=1000, max_depth=8, min_child_weight=6, gamma=0.1, subsample=0.95,
                     colsample_bytree=0.95, reg_alpha=2, objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy',max_depth=1)
ada = AdaBoostClassifier(base_estimator=tree, n_estimators=500, learning_rate=0.1, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(random_state=20)

from sklearn.svm import SVC, LinearSVC
svc = SVC()

from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
```

```python
from sklearn.ensemble import VotingClassifier

eclf = VotingClassifier(estimators=[('forest', forest), ('xgb', xgb1), ('svm', svc), ('logreg', logreg), ('adboost', ada), ('Naive Bayes', gaussian)], voting='hard')
```

```python
#Run this at the time of model training
#eclf.fit(X_train_std, Y_train)

#Run this while predicting from tesst data set
eclf.fit(X, y)
```

```python
eclf.score(X_test_std, Y_test)*100
```

```python
dtest = pd.read_csv('test.csv')
```

```python
dtest['Dependents'] = dtest['Dependents'].str.rstrip('+')
#dtest['Dependents'] = dtest['Dependents'].astype(np.int)
dtest['Gender'] = dtest['Gender'].fillna(dtest['Gender'].dropna().mode().values[0])
dtest['Dependents'] = dtest['Dependents'].fillna( dtest['Dependents'].dropna().mode().values[0])
dtest['Self_Employed'] = dtest['Self_Employed'].fillna( dtest['Self_Employed'].dropna().mode().values[0])
dtest['LoanAmount'] = dtest['LoanAmount'].fillna( dtest['LoanAmount'].dropna().mode().values[0])
dtest['Loan_Amount_Term'] = dtest['Loan_Amount_Term'].fillna( dtest['Loan_Amount_Term'].dropna().mode().values[0])
dtest['Credit_History'] = dtest['Credit_History'].fillna( dtest['Credit_History'].dropna().mode().values[0] )

dtest.reset_index(drop=True, inplace=True)
for i in  range(len(dtest)):
    if dtest['CoapplicantIncome'][i]>0:
        dtest['CoapplicantIncome'][i]=1

dtest.reset_index(drop=True, inplace=True)
dtest.head()
```

```python
list=['Gender','Married','Education','Self_Employed','Self_Employed']
for i in list:
    dtest[i]=label.fit_transform(dtest[i]).astype(int)

columns=['Loan_ID','Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']

dtest=dtest[columns]
dtest.reset_index(drop=True, inplace=True)
print(dtest.info())
dtest.head()
```

```python
x_test = dtest.iloc[:,1:]
x_test= pd.get_dummies(x_test)
x_test.head()
```

```python
x_test_std = slc.transform(x_test)
```

```python
y_test_pred = eclf.predict(x_test_std)
y_test_pred
```

```python
df_final = dtest.drop(['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area'], axis=1)
```

```python
Loan_Status=[]

for i in range(len(y_test_pred)):
    if y_test_pred[i]==int(1):
        Loan_Status.append('Y')
    else:
        Loan_Status.append('N')
```

```python
df_final['Loan_Status']=Loan_Status
```

```python
df_final.to_csv('test_predictions.csv', index=False)
```
